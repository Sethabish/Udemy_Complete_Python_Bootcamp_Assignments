# Milestone Project 1 - Tic Tac Toe

```python
def display(row1,row2,row3):
    print('    1   2   3')
    print('  -------------')
    print('1 | {} | {} | {} |'.format(row1[0],row1[1],row1[2]))
    print('  -------------')
    print('2 | {} | {} | {} |'.format(row2[0],row2[1],row2[2]))
    print('  -------------')
    print('3 | {} | {} | {} |'.format(row3[0],row3[1],row3[2]))
    print('  -------------')
```

```python
def win_check(row1,row2,row3):
    if (row1[0] == 'O' and row1[1] == 'O' and row1[2] == 'O') or \
    (row2[0] == 'O' and row2[1] == 'O' and row2[2] == 'O') or \
    (row3[0] == 'O' and row3[1] == 'O' and row3[2] == 'O') or \
    (row1[0] == 'O' and row2[0] == 'O' and row3[0] == 'O') or \
    (row1[1] == 'O' and row2[1] == 'O' and row3[1] == 'O') or \
    (row1[2] == 'O' and row2[2] == 'O' and row3[2] == 'O') or \
    (row1[0] == 'O' and row2[1] == 'O' and row3[2] == 'O') or \
    (row1[2] == 'O' and row2[1] == 'O' and row3[0] == 'O'):
        return '*** PLAYER "O" WON ***'
    elif (row1[0] == 'X' and row1[1] == 'X' and row1[2] == 'X') or \
    (row2[0] == 'X' and row2[1] == 'X' and row2[2] == 'X') or \
    (row3[0] == 'X' and row3[1] == 'X' and row3[2] == 'X') or \
    (row1[0] == 'X' and row2[0] == 'X' and row3[0] == 'X') or \
    (row1[1] == 'X' and row2[1] == 'X' and row3[1] == 'X') or \
    (row1[2] == 'X' and row2[2] == 'X' and row3[2] == 'X') or \
    (row1[0] == 'X' and row2[1] == 'X' and row3[2] == 'X') or \
    (row1[2] == 'X' and row2[1] == 'X' and row3[0] == 'X'):
        return '*** PLAYER "X" WON ***'
    else:
        pass
```

```python
def x_o(list_x_o):
    if list_x_o[-1] == 'o':
        in_put = input('Player "X" turn: ')
        if in_put == 'x':
            return in_put
        else:
            return 'Wrong Input'
    else:
        in_put = input('Player "O" turn: ')
        if in_put == 'o':
            return in_put
        else:
            return 'Wrong Input'
```

```python
def restart():
    in_put = input('Do you want to play again (y or n): ')
    if in_put == 'y':
        return tic_tac_toe()
    elif in_put == 'n':
        return '*** GAME OVER ***'
    else:
        print('Wrong Input')
        return '*** GAME OVER ***'
```

```python
def tic_tac_toe():
    row1=[' ',' ',' ']
    row2=[' ',' ',' ']
    row3=[' ',' ',' ']
    row_in_1,row_in_2,row_in_3,list_x_o=[],[],[],['o']
    print('*** Welcome to TIC TAC TOE ***')
    display(row1,row2,row3)
    while len(list_x_o)<10:
        block_choice = input('Enter the block choice: ')
        if block_choice == '11' or block_choice == '12' or block_choice == '13':
            if block_choice in row_in_1:
                print('Block already filled, try another...')
            else:
                in_put = x_o(list_x_o)
                if in_put in 'xo':
                    row_in_1.append(block_choice)
                    l = list(block_choice)
                    row1[int(l[1])-1] = in_put.upper()
                    list_x_o.append(in_put)
                else:
                    print('Wrong input')
                display(row1,row2,row3)
        elif block_choice == '21' or block_choice == '22' or block_choice == '23':
            if block_choice in row_in_2:
                print('Block already filled, try another...')
            else:
                in_put = x_o(list_x_o)
                if in_put in 'xo':
                    row_in_2.append(block_choice)
                    l = list(block_choice)
                    row2[int(l[1])-1] = in_put.upper()
                    list_x_o.append(in_put)
                else:
                    print('Wrong Input') 
                display(row1,row2,row3)
        elif block_choice == '31' or block_choice == '32' or block_choice == '33':
            if block_choice in row_in_3:
                print('Block already filled, try another...')
            else:
                in_put = x_o(list_x_o)
                if in_put in 'xo':
                    row_in_3.append(block_choice)
                    l = list(block_choice)
                    row3[int(l[1])-1] = in_put.upper()
                    list_x_o.append(in_put)
                else:
                    print('Wrong Input') 
                display(row1,row2,row3)
        else:
            print('Wrong block choice try again')
        if len(list_x_o)-1<=9:
            if win_check(row1,row2,row3) == '*** PLAYER "X" WON ***' or \
            win_check(row1,row2,row3) == '*** PLAYER "O" WON ***':
                print(win_check(row1,row2,row3))
                return restart()
    else:
        print('*** DRAW ***')
        return restart()
```
